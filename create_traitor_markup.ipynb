# Create Traitor Markup

This notebook creates the traitor md file markup that is used on the main Readme.md file. After I run this process I just copy the markup to the main Readme.md.

```python
import pandas as pd
import io
import requests
```

```python
traitor_df = pd.read_csv("147_traitors.csv", index_col=False)
```

```python
traitor_df
```

```python
traitor_df = traitor_df.fillna('')
```

```python
traitor_df
```

```python
header = 'Photo | Name | State | Chamber | Donors | Twitter | Fired | Demands  \n'
separator = '------------ | ------------ | ------------- | ------------- | ------------- | ------------- | ------------- | -------------\n'

f = open("traitors.md", "w")
f.write(header)
f.write(separator)


for _, traitor_row in traitor_df.iterrows():     

    donor_link = 'n/a'
    if traitor_row.open_secrets_id:
        donor_link = f'[List]({traitor_row.top_donors_link})'
    photo_img = f'<img src="https://github.com/147-Project/147-republican-traitors/blob/master/photos/{traitor_row.picture}?raw=true" width="40" height="40">'
    
    f.write(f'{photo_img} | {traitor_row.full_name} | {traitor_row.state} | {traitor_row.chamber}| {donor_link} | @{traitor_row.twitter} | no | no\n')

f.close()

#open and read the file after the appending:
f = open("traitors.md", "r")
print(f.read())
```
